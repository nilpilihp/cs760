## Q6 hypothesis space

```python
# Decision boundry of D1 model
cols = ['x1', 'x2', 'Outcome']
data = pd.read_csv('data/D1.txt', header=None, names=cols, delimiter=' ')

# define bounds of the domain
min1, max1 = data['x1'].min()-.05, data['x1'].max()+.05
min2, max2 = data['x2'].min()-.05, data['x2'].max()+.05
#interval of 0.05, 110 lines of grid per side
x1grid = np.arange(min1, max1, 0.01) 
x2grid = np.arange(min2, max2, 0.01)
xx, yy = np.meshgrid(x1grid, x2grid)
# 110 x 110 sampling points over domain
r1, r2 = xx.flatten(), yy.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
grid = np.hstack((r1,r2))
xtest = pd.DataFrame(grid,columns=['x1','x2'])
yhat = treeD1.predict(xtest)
zz = yhat.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, zz, cmap='Paired', alpha=0.4)
plt.scatter(x=data['x1'][data['Outcome']==0],y=data['x2'][data['Outcome']==0],label='class 0')
plt.scatter(x=data['x1'][data['Outcome']==1],y=data['x2'][data['Outcome']==1],label='class 1')
plt.legend()
plt.title('Decision Boundary of model trained on D1.txt ')
plt.show()
```

## 7 Learning curve

```python
# Dbig.txt is sorted
cols = ['x1', 'x2', 'Outcome']
data = pd.read_csv('data/Dbig.txt', header=None, names=cols, delimiter=' ')
#permutate
shuffled_df = data.sample(frac=1, random_state=42)
```

```python
n = [32,128,512,2048,8192]
Dtrain_sets = []
models = []
for i,size in enumerate(n):
    data = shuffled_df.iloc[:size]
    Dtrain_sets.append(data)
    models.append(DecisionTree(data,cols[:2]))
Dtest = shuffled_df.iloc[n[-1]:]
```

```python
# show model, total number of nodes shown after construction
for i,model in enumerate(models):
    model()
```

```python
nodes = [4,11,22,51,108]
# error
err = [model.train_error(Dtest)['Error'] for model in models]
err
```

```python
plt.plot(n,err)
plt.title('Testing error vs model trained on different data size')
plt.xlabel('train size')
plt.ylabel('test error')
plt.show()
```

```python
for i in range(5):
    model = models[i]
    data = Dtrain_sets[i]
    # define bounds of the domain
    min1, max1 = data['x1'].min()-.05, data['x1'].max()+.05
    min2, max2 = data['x2'].min()-.05, data['x2'].max()+.05
    #interval of 0.05, 110 lines of grid per side
    x1grid = np.arange(min1, max1, 0.01) 
    x2grid = np.arange(min2, max2, 0.01)
    xx, yy = np.meshgrid(x1grid, x2grid)
    # 110 x 110 sampling points over domain
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
    grid = np.hstack((r1,r2))
    xtest = pd.DataFrame(grid,columns=['x1','x2'])
    yhat = model.predict(xtest)
    zz = yhat.reshape(xx.shape)
    plt.contourf(xx, yy, zz, cmap='Paired', alpha=0.4)
    plt.scatter(x=data['x1'][data['Outcome']==0],y=data['x2'][data['Outcome']==0],label='class 0')
    plt.scatter(x=data['x1'][data['Outcome']==1],y=data['x2'][data['Outcome']==1],label='class 1')
    plt.legend()
    plt.title(f'Decision Boundary of model trained on D{n[i]}')
    plt.show()
```

# Question 3 sklearn

```python
from sklearn import tree 
from sklearn.metrics import accuracy_score
sk_models = []
sk_nodes = []
sk_err=[]
for i in range(5):
    clf = tree.DecisionTreeClassifier().fit(X=Dtrain_sets[i][cols[:2]],y=Dtrain_sets[i]['Outcome'])
    yhat = clf.predict(Dtest[cols[:2]])
    acc = accuracy_score(y_true=Dtest['Outcome'],y_pred=yhat)
    sk_nodes.append(clf.tree_.node_count)
    sk_models.append(clf)
    sk_err.append(round(1-acc,4))
```

```python
print('nodes: ', sk_nodes)   
print('err: ',sk_err)
```

```python
plt.plot(n,sk_err)
plt.title('Testing error vs sklearn model trained on different data size')
plt.xlabel('train size n')
plt.ylabel('test error')
plt.show()
```
