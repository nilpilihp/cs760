```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1.5,1.5,200)
mesh = np.meshgrid(x,x)
norm = np.linalg.norm(mesh,ord=1,axis=0)
plt.scatter(*mesh, c = norm<=1)
plt.title('l1 norm region <=1 shown in yellow')
plt.show()
```

```python
norm = np.linalg.norm(mesh,ord=2,axis=0)
plt.scatter(*mesh, c = norm<=1)
plt.title('l2 norm region <=1 shown in yellow')
plt.show()
```

```python
norm = np.linalg.norm(mesh,ord=np.inf,axis=0)
plt.scatter(*mesh, c = norm<=1)
plt.title('l_inf norm region <=1 shown in yellow')
plt.show()
```

```python
mean = np.array([1,-1])
cov = np.identity(2)*2
x,y = np.random.multivariate_normal(mean, cov, 100).T
```

```python
plt.scatter(x,y)
plt.xlim([-10, 10])
plt.ylim([-10,10])
plt.title(('scatter plot 100 samples from gaussian'))
plt.show()
```

```python
mean_1 = np.array([5,0])
cov_1 = np.array([[1,.25],[.25,1]])
mean_2 = np.array([-5,0])
cov_2 = np.array([[1,-.25],[-.25,1]])

def generator(choice):
  if choice == 0:
    x, y = np.random.multivariate_normal(mean_1, cov_1).T
  elif choice == 1:
    x, y = np.random.multivariate_normal(mean_2, cov_2).T
  return x,y
```

```python
choice = [np.random.choice(np.arange(0, 2), p=[0.7, 0.3]) for i in range(100)]
samples = np.array([generator(c) for c in choice])
```

```python
plt.scatter(samples[:,0],samples[:,1])
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.title('scatter plot 100 samples from mixture gaussian')
plt.show()
```
